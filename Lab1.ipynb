```python
from PIL import Image
import numpy as np
img_path = 'HappyFish.png'
img = Image.open(img_path)
img_np = np.array(img)
R = img_np[:, :, 0]
G = img_np[:, :, 1]
B = img_np[:, :, 2]
grayscale = 0.299 * R + 0.587 * G + 0.114 * B
grayscale_img = Image.fromarray(grayscale.astype(np.uint8))
grayscale_img
```

```python
img_path = 'Lena.jpg'
img = Image.open(img_path)
img_np = np.array(img, dtype = np.float32)
alpha = 2.2
beta = 50
adj_img = alpha * img_np + beta
adj_img = np.clip(adj_img, 0, 255)
adj_img = adj_img.astype(np.uint8)
adj_img = Image.fromarray(adj_img)
adj_img
```

```python
import IPython
from IPython.display import display
import numpy as np
img_1 = 'LinuxLogo.jpg'
img_2 = 'WindowsLogo.jpg'
img_1_img = Image.open(img_1)
img_2_img = Image.open(img_2)
# Resize img_2 to match the size of img_1
img_2_resized = img_2_img.resize(img_1_img.size)
img_1_np = np.array(img_1_img, dtype = np.float32)
img_2_np = np.array(img_2_resized, dtype = np.float32)
frames = []
for alpha in np.arange(0, 1.1, 0.1):
  blended = (1 - alpha) * img_1_np + alpha * img_2_np
  blended = np.clip(blended, 0, 255)
  blended_img = Image.fromarray(blended.astype('uint8'))
  frames.append(blended_img)
frames[0].save('blended_animation.gif', save_all=True, append_images=frames[1:], duration=200, loop=0)
display(IPython.display.Image(filename='blended_animation.gif'))
```
